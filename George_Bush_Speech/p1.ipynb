```python
import spacy
import nltk
from nltk.corpus import state_union
import re
```

```python
txt = state_union.raw('2005-GWBush.txt')
```

```python
print(txt[:500])
```

```python
text = re.sub(r'\(\w+\.\)','',txt)
text = re.sub(r'[^A-Za-z]',' ',text)
i = 4
while i > 0:
    text = re.sub(r" \w "," ",text)
    i -= 1
text = re.sub('\s+',' ',text)
```

```python
print(text[:500])
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
txt = re.sub('\(\w+\.?\)|(-)','',txt)
txt = re.sub('\n',' ',txt)
txt = re.sub('\s+',' ',txt)
```

```python
sentences = nltk.sent_tokenize(txt)
stop_words = nltk.corpus.stopwords.words("english")
word_freq = {}
for w in nltk.word_tokenize(text):
    if w not in stop_words:
        if w not in word_freq.keys():
            word_freq[w.lower()] = 1
        else:
            word_freq[w.lower()] += 1
```

```python
sentences[:20]
```

```python
max_freq = max(word_freq.values())
for k in word_freq.keys():
    word_freq[k] = word_freq[k] / max_freq
```

```python
x = 100
for k,v in word_freq.items():
    print(k,'-->',v)
    if x == 0:
        break
    x -= 1
```

```python
sen_scores = {}
for sen in sentences:
    for word in nltk.word_tokenize(sen.lower()):
        if word in word_freq.keys():
                if sen not in sen_scores.keys():
                    sen_scores[sen] = word_freq[word]
                else:
                    sen_scores[sen] += word_freq[word]
```

Summarizing Geore Bush Speech 

```python
import heapq
summary_sentences = heapq.nlargest(5, sen_scores, key=sen_scores.get)
summary = ' '.join(summary_sentences)
print(summary)
```
